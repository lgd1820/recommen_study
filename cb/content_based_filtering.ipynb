```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset/tmdb_5000_movies.csv")
df.head()
```

```python
column_list = ["id", "genres", "vote_average", "vote_count", "popularity", "title", "keywords", "overview"]
data = df[column_list]
data = data.dropna(axis=0)
data.shape
```

```python
data.head()
```

```python
data.describe()
```

```python
sns.kdeplot(data["vote_average"], )
plt.title("vote average")
plt.show()
```

```python
# 평점 전처리
m = data['vote_count'].quantile(0.9)
print(m)
quantile_data=data.loc[data['vote_count'] >= m]
```

```python
quantile_data.shape
```

```python
def weight_rating(x, m, c):
    v = x["vote_count"]
    r = x["vote_average"]
    
    return (v / (v + m) * r) + (m / (v + m) * c)
```

```python
m = data['vote_count'].quantile(0.9)
c = quantile_data['vote_average'].mean()
quantile_data['score'] = quantile_data.apply(weight_rating, axis=1, args=(m,c))
quantile_data.head()
```

```python
import json
quantile_data['keywords'] = quantile_data['keywords'].apply(lambda x: " ".join([genre['name'] for genre in json.loads(x)]))
quantile_data['keywords'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

counter_vector = CountVectorizer(ngram_range=(1,3))
counter_vector_keywords = counter_vector.fit_transform(quantile_data['keywords'])
similarity_keywords = cosine_similarity(counter_vector_keywords, counter_vector_keywords).argsort()[:,::-1]
similarity_keywords
```

```python
def recommend_movie_list(quantile_data, movie_title, similarity_keywords, top=30):
    target_movie_index = quantile_data[quantile_data["title"] == movie_title].index.values
    sim_index = similarity_keywords[target_movie_index, :top].reshape(-1)
    # sim_index = sim_index[sim_index != target_movie_index]
    result = quantile_data.iloc[sim_index][:10]
    
    return result

recommend_movie_list(quantile_data, "Avatar", similarity_keywords)
```
